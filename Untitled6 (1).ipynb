```python
df = df.drop(['waferIndex'], axis = 1)
```

```python
#wafermap 의 차원을 나타내는 waferDim을 dataframe에 추가
def find_dim(x):
    dim0=np.size(x,axis=0)
    dim1=np.size(x,axis=1)
    return dim0,dim1
df['waferMapDim']=df.waferMap.apply(find_dim)
df.sample(5)
```

```python
#차원의 최대와 최소 확인
max(df.waferMapDim), min(df.waferMapDim)
```

```python
#중복되는 고유 차원값의 개수 확인
uni_waferDim=np.unique(df.waferMapDim, return_counts=True)
uni_waferDim[0].shape[0]
```

```python
#각각의 class에 대해서 failureNum에 index 0~8으로 저장
#training 과 test data에 대해 분류
df['failureNum']=df.failureType
df['trainTestNum']=df.trianTestLabel
mapping_type={'Center':0,'Donut':1,'Edge-Loc':2,'Edge-Ring':3,'Loc':4,'Random':5,'Scratch':6,'Near-full':7,'none':8}
mapping_traintest={'Training':0,'Test':1}
df=df.replace({'failureNum':mapping_type, 'trainTestNum':mapping_traintest})
```

```python
tol_wafers = df.shape[0]
tol_wafers
```

```python
#label이 있는 data, 유효한 label이 있는 data, label이 none인 data로 분류
df_withlabel = df[(df['failureNum']>=0) & (df['failureNum']<=8)]
df_withlabel =df_withlabel.reset_index()
df_withpattern = df[(df['failureNum']>=0) & (df['failureNum']<=7)]
df_withpattern = df_withpattern.reset_index()
df_nonpattern = df[(df['failureNum']==8)]
df_withlabel.shape[0], df_withpattern.shape[0], df_nonpattern.shape[0]
```

```python
# 'waferMapDim'이 (25,27)인 데이터만 선택
df_selected_dim = df_withlabel[df_withlabel['waferMapDim'] == (25,27)]

# 'failureNum'의 종류에 따른 개수를 출력
failureNum_counts = df_selected_dim['failureNum'].value_counts()
for failure_type, count in failureNum_counts.items():
    type_name = next((key for key, value in mapping_type.items() if value == failure_type), None)
    print(f"failureNum: {type_name}, 데이터 개수: {count}")
```

```python
df_train = df_selected_dim[(df_selected_dim.trainTestNum==0)]
df_test = df_selected_dim[(df_selected_dim.trainTestNum==1)]

df_train.shape ,df_test.shape
```

```python
X_train, y_train= np.array(df_train['waferMap'].values.tolist()), np.array(df_train["failureNum"])

X_test, y_test= np.array(df_test['waferMap'].values.tolist()), np.array(df_test["failureNum"])
```

```python
X_train.shape, y_train.shape , X_test.shape, y_test.shape
```

```python
unique_classes,counts =np.unique(y_train, return_counts=True)
for cls, count in zip(unique_classes, counts):
    print(f"{cls}: {count} samples")
```

```python
#0,1,2로 one-hot-encoding
from tensorflow.keras.utils import to_categorical

X_train = to_categorical(X_train, num_classes=3)
X_test = to_categorical(X_test, num_classes=3)
```

```python
X_train.shape,X_test.shape
```

```python
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, img_to_array
import matplotlib.pyplot as plt
# 이미지 데이터 증강을 위한 ImageDataGenerator 생성
data_gen = ImageDataGenerator(vertical_flip=True, horizontal_flip=True, rotation_range=10)

for i in range(8):
    # y_train이 1인 경우 건너뛰기
    if i == 1:
        continue

    label_indices = np.where(y_train == i)[0]
    X_train_label = X_train[label_indices]
    time = int((14000 / len(label_indices))) - 1

    # 증강된 이미지를 담을 리스트
    augmented_images = []

    # 특정 레이블에 대한 이미지 데이터를 가져와서 증강
    for img_array in X_train_label:
        img_array = img_array.reshape((1,) + img_array.shape)  # 모델이 받아들일 수 있는 형태로 변환

        # flow 메서드를 사용하여 이미지 증강 수행
        augmented_data = data_gen.flow(img_array, batch_size=1)

        # 증강된 이미지를 리스트에 추가
        augmented_images.extend([next(augmented_data)[0] for _ in range(time)])

    X_train_new = np.array(augmented_images)
    y_train_new = np.full((X_train_new.shape[0],), i)
    X_train = np.concatenate((X_train, X_train_new), axis=0)
    y_train = np.concatenate((y_train, y_train_new), axis=0)

```

```python
unique_classes,counts =np.unique(y_train, return_counts=True)
for cls, count in zip(unique_classes, counts):
    print(f"{cls}: {count} samples")
```

```python
#data shuffling
indices = np.arange(len(X_train))
np.random.shuffle(indices)

X_train = X_train[indices]
y_train = y_train[indices]
```

```python
import os
from os.path import join

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
import keras
from keras import layers, Input, models
from keras.utils import to_categorical
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from sklearn.model_selection import train_test_split
```

```python
X_train = tf.convert_to_tensor(X_train,dtype=tf.float32)
y_train = tf.convert_to_tensor(y_train,dtype=tf.float32)

X_test = tf.convert_to_tensor(X_test,dtype=tf.float32)
y_test = tf.convert_to_tensor(y_test,dtype=tf.float32)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam

import tensorflow as tf
from tensorflow.keras import layers, models

# 클래스 개수
num_classes = 9

# 모델 정의
model = models.Sequential()

# 첫 번째 합성곱 층
model.add(layers.Conv2D(16, (3, 3), activation='relu', padding='same', input_shape=(25, 27, 3)))

# 두 번째 합성곱 층
model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(layers.MaxPooling2D((1, 1)))

# 세 번째 합성곱 층
model.add(layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(layers.MaxPooling2D((2, 2)))

# 네 번째 합성곱 층 (추가)
model.add(layers.Conv2D(192, (3, 3), activation='relu', padding='same'))

# 다섯 번째 합성곱 층 (추가)
model.add(layers.Conv2D(256, (3, 3), activation='relu', padding='same'))

# 완전 연결 층
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))

# 출력 층
model.add(layers.Dense(num_classes, activation='softmax'))

# 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 모델 구조 출력
model.summary()


```

```python
# Convert labels to one-hot encoding
y_train_encoded = to_categorical(y_train, num_classes=num_classes)

```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2, batch_size=64)

```

```python
# Access accuracy values from the training history
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Plot accuracy over epochs
epochs = range(1, len(train_accuracy) + 1)
plt.plot(epochs, train_accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'r', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```
